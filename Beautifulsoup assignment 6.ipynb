```python
#Iporting required liberaries
from bs4 import BeautifulSoup
import requests 
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/womens/team-rankings/odi")
page
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
team=soup.find_all('span',class_='u-hide-phablet')
team[:10]
```

```python
cric_team=[]
for i in team:
    cric_team.append(i.text)
cric_team[:10]
```

```python
points=soup.find('td',class_='rankings-block__banner--points')
points
```

```python
cric_points=[]
for i in points:
    cric_points.append(i)
cric_points
```

```python
points1=soup.find_all('td',class_='table-body__cell u-center-text')
points1[1:18:2]
```

```python
cric_points1=[]
for i in points1:
    cric_points1.append(i.text)
cric_points1[1:18:2]
```

```python
cric_points.extend(cric_points1[1:18:2])
cric_points
```

```python
matches=soup.find('td',class_='rankings-block__banner--matches')
matches
```

```python
cric_matches=[]
for i in matches:
    cric_matches.append(i)
cric_matches
```

```python
matches1=soup.find_all('td',class_='table-body__cell u-center-text')
matches1[:9]
```

```python
cric_matches1=[]
for i in matches1:
    cric_matches1.append(i.text)
cric_matches1[1:18:2]
```

```python
cric_matches.extend(cric_matches1[1:18:2])
cric_matches
```

```python
ratings=soup.find('td',class_='rankings-block__banner--rating u-text-right')
ratings
```

```python
cric_ratings=[]
for i in ratings:
    cric_ratings.append(i)
cric_ratings
```

```python
ratings1=soup.find_all('td',class_='table-body__cell u-text-right rating')
ratings1
```

```python
cric_ratings1=[]
for i in ratings1:
    cric_ratings1.append(i.text)
cric_ratings1
```

```python
cric_ratings.extend(cric_ratings1[:9])
cric_ratings
```

```python
import pandas as pd
cric=pd.DataFrame({})
cric['teams']=cric_team[:10]
cric['points']=cric_points[:10]
cric['ratings']=cric_ratings[:10]
cric['matches']=cric_matches[:10]
```

```python
cric
```
