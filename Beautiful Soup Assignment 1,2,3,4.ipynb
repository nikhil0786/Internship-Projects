# python program to display all the header tags from ‘en.wikipedia.org/wiki/Main_Page’.

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
title=soup.find_all('h2',class_='mp-h2')
```

```python
title
```

```python
title.text
```

```python
#now we will extract the text from these tags one by one by looping over these tags
wiki_title=[]
for i in title:
    wiki_title.append(i.text)
wiki_title
```

```python
import pandas as pd
wiki=pd.DataFrame({})
wiki['title']=wiki_title
```

```python
wiki
```

# python program to display IMDB’s Top rated 100 movies’ 

```python
page1=requests.get("https://www.imdb.com/chart/top/")
page1
```

```python
soup=BeautifulSoup(page1.content,'html.parser')
soup
```

```python
titles=soup.find_all('td',class_='titleColumn')
titles
```

```python
imdb_title=[]
for i in titles:
    imdb_title.append(i.text.replace('\n',''))
imdb_title[0:100]
```

```python
ratings=soup.find_all('td',class_='ratingColumn imdbRating')
ratings
```

```python
imdb_ratings=[]
for i in ratings:
    imdb_ratings.append(i.text.replace('\n',''))
imdb_ratings[0:100]
```

```python
years=soup.find_all('span',class_='secondaryInfo')
years
```

```python
imdb_years=[]
for i in years:
    imdb_years.append(i.text.replace('\n',''))
imdb_years[0:100]
```

```python
import pandas as pd
imdb=pd.DataFrame({})
imdb['titles']=imdb_title[:100]
imdb['ratings']=imdb_ratings[:100]
imdb['years']=imdb_years[:100]
```

```python
imdb
```

# python program to scrap book name, author name, genre and book review of any 5 books from 
‘www.bookpage.com’

```python
page3=requests.get("https://bookpage.com/search/?utf8=%E2%9C%93&query=top+5+books")
page3
```

```python
soup=BeautifulSoup(page3.content,'html.parser')
soup
```

```python
bname=soup.find_all('h4',class_='italic')
bname[:5]
```

```python
book_name=[]
for i in bname:
    book_name.append(i.text.replace('\n',''))
book_name[0:5]
```

```python
aname=soup.find_all('p',class_='sans bold')
aname[:5]
```

```python
author_name=[]
for i in aname:
    author_name.append(i.text.replace('\n',''))
author_name[0:5]
```

```python
bgenre=soup.find_all('p',class_='genre-links hidden-phone')
bgenre[:5]
```

```python
book_genre=[]
for i in bgenre:
    book_genre.append(i.text.replace('\n',''))
book_genre[0:5]
```

```python
breview=soup.find_all('p',class_='excerpt')
breview[:5]
```

```python
book_review=[]
for i in breview:
    book_review.append(i.text.replace('\n',''))
book_review[0:5]
```

```python
bookpage=pd.DataFrame({})
bookpage['Name']=book_name[:5]
bookpage['author']=author_name[:5]
bookpage['genre']=book_genre[:5]
bookpage['review']=book_review[:5]
```
