```python
#Iporting required liberaries
from bs4 import BeautifulSoup
import requests 
```

```python
page=requests.get("https://www.icc-cricket.com/rankings/mens/team-rankings/odi")
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
team=soup.find_all('span',class_='u-hide-phablet')
team[:10]
```

```python
#now we will extract the text from these tags one by one by looping over these tags
cric_team=[]
for i in team:
    cric_team.append(i.text)
cric_team[:10]
```

```python
matches1=soup.find_all('td',class_='rankings-block__banner--matches')
```

```python
matches1
```

```python
cric_matches1=[]
for i in matches1:
    cric_matches1.append(i.text)
cric_matches1
```

```python
cric_matches1
```

```python
matches=soup.find_all('td',class_='table-body__cell u-center-text')
```

```python
matches[0:20:2]
```

```python
cric_matches=[]
for i in matches:
    cric_matches.append(i.text)
cric_matches[:19:2]
```

```python
cric_matches1.extend(cric_matches[:18:2])
```

```python
cric_matches1
```

```python
points1=soup.find_all('td',class_='rankings-block__banner--points')
points1
```

```python
cric_points1=[]
for i in points1:
    cric_points1.append(i.text)
cric_points1
```

```python
points=soup.find_all('td',class_='table-body__cell u-center-text')
points[1:20:2]
```

```python
cric_points=[]
for i in points:
    cric_points.append(i.text)
cric_points[1:20:2]
```

```python
cric_points1.extend(cric_points[1:18:2])
```

```python
cric_points1
```

```python
ratings1=soup.find_all('td',class_='rankings-block__banner--rating u-text-right')
ratings1
```

```python
cric_ratings1=[]
for i in ratings1:
    cric_ratings1.append(i.text.replace('\n',''))
cric_ratings1
```

```python
ratings=soup.find_all('td',class_='table-body__cell u-text-right rating')
ratings
```

```python
cric_ratings=[]
for i in ratings:
    cric_ratings.append(i.text.replace('\n',''))
cric_ratings[:10]
```

```python
cric_ratings1.extend(cric_ratings)
```

```python
cric_ratings1[:10]
```

```python
import pandas as pd
cric=pd.DataFrame({})
cric['teams']=cric_team[:10]
cric['matches']=cric_matches1[:10]
cric['points']=cric_points1[:10]
cric['ratings']=cric_ratings1[:10]
```
